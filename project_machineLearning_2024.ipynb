### Load CSVs into DataFrames

```python
import pandas as pd

#Load KenPom Barttorvik dataset into a dataframe
kenpom = pd.read_csv('KenPom Barttorvik.csv')

#Load Preseason Votes dataset into a dataframe
preseason = pd.read_csv('Preseason Votes.csv')

#Load Team Results dataset into a dataframe
team = pd.read_csv('Team Results.csv')

#Load Tournament Matchups dataset into a dataframe
matchups = pd.read_csv('Tournament Matchups.csv')
```

### Merge All CSVs Together

```python
# Left join KenPom and Preseason based on TEAM (team name) and YEAR
first_merged_df = pd.merge(kenpom, preseason, on=['TEAM', 'YEAR'], how='left')

#Result (Makes sure all teams are loaded) (68 teams for 2024)
first_merged_df.head(100)
```

```python
# Left join the KenPom/Preseason dataset and Team Dataset on TEAM (team name)
second_merged_df = pd.merge(first_merged_df, team, on='TEAM', how='left')

#Result (Makes sure all teams are loaded) (68 teams for 2024)
second_merged_df.head(100)
```

```python
second_merged_df.shape
```

```python
# Left join the merged dataset and the matchups dataset on TEAM (team name)
final_merged_df = pd.merge(second_merged_df, matchups, on=['TEAM', 'YEAR'], how='right')

#Result (Makes sure all teams are loaded) (68 teams for 2024)
final_merged_df.head(100)
```
